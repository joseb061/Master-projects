```python
lambda_data1 = pd.read_csv('lambda_resuts.txt',sep=";",header=None)
lambda_data1.columns = ["Lambda","Distance","Number of iterations","Execution time A*"]
```

```python
# Selecting lambda that minimizes execution time
data = lambda_data1[["Lambda","Execution time A*"]].loc[lambda_data1['Distance']==958.765625]
data[data['Execution time A*'] == data['Execution time A*'].min()] 
```

```python
lambda_data1.loc[lambda_data1["Lambda"]==1]
lambda_data1.loc[lambda_data1["Lambda"]==1.12]
```

```python
graph1 = seaborn.relplot(
    data=lambda_data1, kind="line",
    x="Lambda", y="Number of iterations",
)
graph1.fig.suptitle('Number of iterations by lambda value')
plt.savefig("Iterations.png")
```

```python
graph2 = seaborn.relplot(
    data=lambda_data1, kind="line",
    x="Lambda", y="Distance",
)
graph2.fig.suptitle('Distance by lambda value')
plt.savefig("Distance.png")
```

```python
graph3 = seaborn.relplot(
    data=lambda_data1, kind="line",
    x="Lambda", y="Execution time A*",
)
graph3.fig.suptitle('Execution time by lambda value')
plt.savefig("Execution_time.png")
```

```python
import matplotlib.pyplot as plt
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(lambda_data1['Lambda'], lambda_data1['Execution time A*'], color="red")
# set x-axis label
ax.set_xlabel("Lambda",fontsize=14)
# set y-axis label
ax.set_ylabel("Execution time A*",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(lambda_data1['Lambda'], lambda_data1['Distance'],color="blue")
ax2.set_ylabel("Distance",color="blue",fontsize=14)
#plt.show()
plt.savefig("Exe_vs_distance.png")
```
