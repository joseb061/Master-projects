```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
fitnes = pd.read_csv('/home/joseba/Master/Master-git/Optimisation/Genetic-Algorithms/GA/results/fitnes_results.txt', sep=" ", header=None)
params = pd.read_csv('/home/joseba/Master/Master-git/Optimisation/Genetic-Algorithms/GA/results/params_resuts.txt', sep=" ", header=None)
years = pd.read_csv('/home/joseba/Master/Master-git/Optimisation/Genetic-Algorithms/GA/results/years_resuts.txt', sep=" ", header=None)
fitnes.columns = ["fitnes"]
years.columns = ["2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017"]
reals = [15329., 14177., 13031., 9762, 11271, 8688., 7571., 6983., 4778., 2067., 1586., 793.]
```

```python
plt.figure(figsize=(7, 5))
plt.title('Fitness function')
plt.plot(fitnes.index,fitnes["fitnes"], lw=2, label = 'Population fitness')
plt.ylim(1800000, 6000000)

plt.legend()
plt.xlabel("Iteration number")
plt.ylabel("Fitness value")
plt.savefig('total_fitnes.png')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.title('Fitness function')
plt.plot(fitnes.index,fitnes["fitnes"], lw=2, label = 'Population fitness')
plt.ylim(1800000, 1900000)
plt.xlim(1000,5000)
plt.vlines(x = 3210, ymin = 1810000, ymax = 1860000, colors = 'purple', linestyles='dashdot',
                                    label = 'max. iter. number')

plt.legend()

plt.xlabel("Iteration number")
plt.ylabel("Fitness value")
plt.savefig('partial_fitnes.png')
plt.show()
```

```python
transpose = years.T
transpose["years"] = transpose.index

compare_years = pd.DataFrame() 
compare_years["years"] = transpose["years"]
compare_years["predicted"] = transpose.iloc[:, -2]
compare_years["real"] = reals
```

```python
plt.figure(figsize=(8, 5))
plt.title('Results by year')
plt.plot(compare_years["years"], compare_years["predicted"], color="blue", label = "predicted values")
plt.plot(compare_years["years"], compare_years["real"],color="orange", label = "real values")

plt.legend()
plt.xlabel("Years")
plt.ylabel("x(t)")

plt.ylabel("Fitnes value")
plt.savefig('years_fitnes.png')
plt.show()
```
