```python
import matplotlib.pyplot as plt
import numpy as np
```

# Ronsebroc's function 3D surface plot

```python

def f(x, y):
    return (np.array(100*(Y-X**2)**2 + (1-X)**2))
    

x = np.linspace(-3.7, 3.7, 60)
y = np.linspace(-2.7, 10.7, 60)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(12,10))
fig.subplots_adjust(top=1.1, bottom=-.1) #centrar la imagen
ax = plt.axes(projection='3d')

ax.view_init(40, 25)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='plasma', edgecolor='none', alpha = 1)

ax.set_xlabel('x')
ax.set_ylabel('y')
```

# 3D and 2D of the Gradient trajectory

```python
x, y, z = np.loadtxt('data/gradient_resuts.txt', delimiter=' ', unpack=True)
x_mod, y_mod, z_mod = np.loadtxt('data/gradient_resuts_modif.txt', delimiter=' ', unpack=True)

fig = plt.figure(figsize=(6,5))

left, bottom, width, height = 0.1, 0.1, 1.4, 1.4
ax = fig.add_axes([left, bottom, width, height]) 
start, stop, n_values = -2.7, 3.7, 800

x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.array(100*(Y-X**2)**2 + (1-X)**2)

plt.contour(X, Y, Z, 80 ,cmap='plasma')
plt.plot(x, y, label=u'Search path real xo', color='red', linewidth=2)
plt.plot(x_mod, y_mod, label=u'Search path arbitrary xo', color='green', linewidth=2)
plt.plot(1, 1, 'or', label=u'Absolute minimum', color="blue")
#plt.colorbar()
plt.legend(loc='best')

#ax.set_title('Rosenbrock’s function', fontsize=10)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y', fontsize=20)




plt.show()
```

# Print a Graph in the area where the function is non-convex

```python
def f(x, y):
    return (np.array( 200*(6*X**4-6*X**2*Y+Y**2) + 2*X**2))

fig = plt.figure(figsize=(6,5))

left, bottom, width, height = 0.1, 0.1, 1.4, 1.4
ax = fig.add_axes([left, bottom, width, height]) 
start, stop, n_values = -1.7, 1.7, 800

x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X,Y)

plt.contour(X, Y, Z*(Z<0), 80 ,cmap='plasma')
plt.plot(1, 1, 'or', label=u'Absolute minimum', color="red")
plt.colorbar()
plt.legend(loc='best')

#ax.set_title('Rosenbrock’s function', fontsize=10)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y', fontsize=20)




plt.show()
```

# Print a Graph in the area where the function is Convex

```python
def f(x, y):
    return (np.array( 200*(6*X**4-6*X**2*Y+Y**2) + 2*X**2))

fig = plt.figure(figsize=(6,5))

left, bottom, width, height = 0.1, 0.1, 1.4, 1.4
ax = fig.add_axes([left, bottom, width, height]) 
start, stop, n_values = -10.7, 10.7, 800

x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X,Y)

plt.contour(X, Y, Z*(Z>=0), 80 ,cmap='plasma')
plt.plot(1, 1, 'or', label=u'Absolute minimum', color="red")
plt.colorbar()
plt.legend(loc='best')
plt.axhline(y = 1., color = 'r', linestyle = '-')

#ax.set_title('Rosenbrock’s function', fontsize=10)
ax.set_xlabel('x', fontsize=20)
ax.set_ylabel('y', fontsize=20)




plt.show()
```
