```python

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Spearman

```python
df = pd.read_csv("Spearman.csv")

df[df["wave_vec"] == "second"].sort_values(["correlation_vec_temp"])
```

```python

second_wave = df[df["wave_vec"] == "second"].sort_values(["correlation_vec_temp"], ascending=False)
second_wave.reindex(second_wave["provincias_vec"])
pastel_colors = sns.color_palette('pastel')[0:5]




fig = plt.figure(figsize=(15, 10))
ax = fig.add_axes([0,0,1,1])
provinces = second_wave["provincias_vec"]
correlations = second_wave["correlation_vec_temp"]
ax.barh(provinces,correlations, color=pastel_colors)
plt.show()
```
