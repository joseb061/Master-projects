```python
import pandas as pd
import seaborn  
```

```python
lambda_data = pd.read_csv('lambda_resuts.txt',sep=";",header=None)
```

```python
lambda_data.columns = ["Lambda","Distance","Number of iterations","Execution time A*"]
```

```python
seaborn.relplot(
    data=lambda_data, kind="line",
    x="Lambda", y="Number of iterations",
)
```

```python
seaborn.relplot(
    data=lambda_data, kind="line",
    x="Lambda", y="Distance",
)
```

```python
seaborn.relplot(
    data=lambda_data, kind="line",
    x="Lambda", y="Execution time A*",
)
```
