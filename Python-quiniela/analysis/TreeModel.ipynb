Importing the data. We only care about the final standings of the 3 last seasons and of the match day standings of last season

```python
# Creating a result column for LastSeasonsStandings
def result(row):
    aux = row['last_results']
    return aux[2] # Me los cuenta como caracteres
```

```python
SeasonStanding = pd.read_excel(r'..\reports\SeasonStanding.xlsx',header=0,index_col=False)
SeasonStanding = SeasonStanding.drop('Unnamed: 0', axis=1)
LastSeasonsStandings = SeasonStanding.loc[(SeasonStanding['season']=='2020-2021')|(SeasonStanding['season']=='2019-2020')|(SeasonStanding['season']=='2018-2019')|(SeasonStanding['season']=='2017-2018')]

MatchdayStanding = pd.read_excel(r'..\reports\MatchdayStanding.xlsx',header=0,index_col=False)
MatchdayStanding = MatchdayStanding.drop('Unnamed: 0', axis=1)
LastMatchdayStandings = MatchdayStanding.loc[(MatchdayStanding['season']=='2019-2020')]
```

```python
LastMatchdayStandings['result'] = LastMatchdayStandings.apply(lambda row: result(row), axis=1)
for col in ['last_results','result','team',]:
    LastMatchdayStandings[col] = LastMatchdayStandings[col].astype('category')
    LastMatchdayStandings[f'{col}_encoded'] = LastMatchdayStandings[col].cat.codes

LastMatchdayStandings.head(20)
```
